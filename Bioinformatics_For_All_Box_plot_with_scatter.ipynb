```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import itertools

# Example data
# Creating a DataFrame with random data for 3 groups
np.random.seed(42)
data = {
    'Group': np.repeat(['A', 'B', 'C'], 30),
    'Value': np.concatenate([np.random.normal(5, 1, 30),
                             np.random.normal(6, 1, 30),
                             np.random.normal(7, 1, 30)])
}
df = pd.DataFrame(data)

# Function to perform pairwise statistical tests (e.g., Mann-Whitney U test or t-test)
def pairwise_tests(df, group_column, value_column, test_type='t-test'):
    results = []
    groups = df[group_column].unique()
    for group1, group2 in itertools.combinations(groups, 2):
        group1_data = df[df[group_column] == group1][value_column]
        group2_data = df[df[group_column] == group2][value_column]

        if test_type == 't-test':
            stat, pval = stats.ttest_ind(group1_data, group2_data)
            test_name = 't-test'
        elif test_type == 'mannwhitneyu':
            stat, pval = stats.mannwhitneyu(group1_data, group2_data)
            test_name = 'Mann-Whitney U'
        else:
            raise ValueError('Unsupported test type. Use "t-test" or "mannwhitneyu".')

        results.append({
            'Group1': group1,
            'Group2': group2,
            'Stat': stat,
            'P-value': pval,
            'Test': test_name
        })

    return pd.DataFrame(results)

# Function to create box plot with scatter
def plot_box_with_scatter(df, group_column, value_column, title='Box plot with scatter'):
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=group_column, y=value_column, data=df, showfliers=False, width=0.6, palette='Set2')
    sns.stripplot(x=group_column, y=value_column, data=df, jitter=True, color='black', size=5, alpha=0.7)
    plt.title(title)
    plt.show()

# Perform pairwise statistical tests (e.g., t-test)
test_results = pairwise_tests(df, 'Group', 'Value', test_type='t-test')

# Display the test results
print(test_results)

# Create box plot with scatter
plot_box_with_scatter(df, 'Group', 'Value', title='Box Plot with Scatter and Pairwise T-tests')

```
